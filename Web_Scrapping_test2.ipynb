```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://tradingeconomics.com/commodity/copper"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
get_title = soup.title
print(get_title)
```

```python
print(soup.prettify())
```

```python
my_tables = soup.find('table', {'class':'table table-hover table-striped'})
#print(my_table)
my_tables
```

```python
data = []
my_tables = soup.find('table', {'class':'table table-hover table-striped'})
table_body = my_tables.find('tbody')
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
df.head(100)
```

```python
header = []
my_tables2 = soup.find('table', {'class':'table table-hover table-striped'})
table_body2 = my_tables2.find('thead')
rows2 = table_body2.find_all('tr')
for row in rows2:
    cols2 = row.find_all('th')
    cols2 = [ele.text.strip() for ele in cols2]
    header.append([ele for ele in cols2 if ele])
```

```python
print(header)
```
